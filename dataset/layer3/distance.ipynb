```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('yearWise.csv')
```

```python
dist = np.zeros([7102, 7102], dtype=float)
```

```python
df.set_index("id", inplace = True)
```

```python
for i in range(7102):
	dist[i][i] = 1
	# print(i)
	i_list = df.loc[i].to_list()
	# break
	for j in range(i+1, 425):
		j_list = df.loc[j].to_list()
		# print(j_list)
		my_rho = np.corrcoef(i_list, j_list)[0][1]
		# print(my_rho[0][1])
		dist[i][j] = my_rho
		dist[j][i] = my_rho
		# break
	# break
```

```python
df_write = pd.DataFrame(data = dist, columns=[i for i in range(0, 7102)])
```

```python
df_write.to_csv("pearson_distance.csv")
```
